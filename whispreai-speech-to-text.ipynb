```python
import whisper
import gradio as gr

def transcribe(audio):
    model = whisper.load_model("base")
    result = model.transcribe(audio)
    return result["text"]

gr.Interface(fn=transcribe, inputs="file", outputs="text").launch(debug=True, share=True)
```
